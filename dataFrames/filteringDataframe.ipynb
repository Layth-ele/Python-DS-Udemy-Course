# IMPORTING THE FULL DATA FRAMES

```python
stats = pd.read_csv('Demographic-Data.csv')
```

```python
#RENAMING COLUMS OF A DATA FRAME TO BE CONSISTENT

stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers',
       'IncomeGroup']

stats.head(4)
```

# FILTERING

```python
filter1 = stats.InternetUsers < 2

filter1
```

```python
filter2 = stats.BirthRate > 40
filter2
```

```python
stats[filter1]
```

```python
stats[filter2]
```

```python
stats[filter1 & filter2] 
```

```python
stats[(stats.BirthRate > 40) & (stats.InternetUsers < 2)]
```

```python
stats[stats.IncomeGroup == 'High income']
```

```python
# HOW TO GET UNIQUE CATEGORIES
stats.IncomeGroup.unique()
```
