# IMPORTING THE FULL DATA FRAMES

```python
stats = pd.read_csv('Demographic-Data.csv')
```

```python
#RENAMING COLUMS OF A DATA FRAME TO BE CONSISTENT

stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers',
       'IncomeGroup']

stats.head(4)
```

# IMPORT SEABORN

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,8))
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
sns.set(rc = {'figure.figsize':(14,6)})
%matplotlib inline
```

# Distribution:

```python
vis1 = sns.displot(x = 'InternetUsers', data = stats, bins = 30,aspect=2)
```

# Boxplots:

```python
plt.figure(figsize = (12,6))
vis2 = sns.boxenplot(x = 'IncomeGroup', y = 'BirthRate', data = stats)
```

# LINEPOT:

```python
plt.figure(figsize = (12,6))
vis3 = sns.lineplot(data = stats)
```

# SEABORN GALLERY :

```python
#Chech Seaborn gallery for more gallery examples
plt.figure(figsize = (12,6))
vis4 = sns.violinplot(x = 'IncomeGroup', y = 'BirthRate', data = stats)
```

```python
vis5 = sns.pairplot(data = stats, size= 4)
```

```python
plt.figure(figsize = (12,6))
vis6 = sns.barplot(x = 'IncomeGroup', y = 'BirthRate', data = stats)
```

```python
vis6 = sns.lmplot(x = 'InternetUsers', y = 'BirthRate', data = stats,\
                  fit_reg = False, hue='IncomeGroup',  aspect = 2 ,  )
```

# KEYWORD ARGUMENTS IN PYTHON

```python
vis6 = sns.lmplot(x = 'InternetUsers', y = 'BirthRate', data = stats,\
                  fit_reg = False, hue='IncomeGroup',  aspect = 2 ,  \
                 scatter_kws = {'s': 150})
```
