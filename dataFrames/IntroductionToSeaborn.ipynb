# IMPORTING THE FULL DATA FRAMES

```python
stats = pd.read_csv('Demographic-Data.csv')
```

```python
#RENAMING COLUMS OF A DATA FRAME TO BE CONSISTENT

stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers',
       'IncomeGroup']

stats.head(4)
```

# IMPORT SEABORN

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,8))
import seaborn as sns
%matplotlib inline
```

# Distribution:

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.set_theme(style="darkgrid")
vis1 = sns.displot(x = 'InternetUsers', data = stats, bins = 30)
```

# Boxplots:

```python
vis2 = sns.boxenplot(x = 'IncomeGroup', y = 'BirthRate', data = stats )
```

# LINEPOT:

```python
vis3 = sns.lineplot(data = stats)
```

# SEABORN GALLERY :

```python
#Chech Seaborn gallery for more gallery examples

vis4 = sns.violinplot(x = 'IncomeGroup', y = 'BirthRate', data = stats)
```

```python
vis5 = sns.pairplot(data = stats)
```

```python
vis6 = sns.barplot(x = 'IncomeGroup', y = 'BirthRate', data = stats,
            label="Total", color="b")
```
