# Movie Rating Jointplot

```python
#importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 

warnings.filterwarnings('ignore')

%matplotlib inline 
```

```python
# Manipulation dataset as previous section

movies = pd.read_csv('Movie-Ratings.csv')
movies.columns = ['Film', 'Genre','CriticRating', 'AudienceRatings'\
              ,'BudgetMillions','Year']
movies.Film = movies.Film.astype('category')
movies.Genre = movies.Genre.astype('category')
movies.Year = movies.Year.astype('category')
movies.head()
```

<h3>JointPlot</h3>

```python
# Joinplot scatter kind property and its a default value

J = sns.jointplot(data= movies ,x= 'CriticRating' ,y= 'AudienceRatings')
```

```python
# Joinplot hex kind property 
JHex = sns.jointplot(data= movies ,x= 'CriticRating' ,\
                     y= 'AudienceRatings' , kind= 'hex')
```
