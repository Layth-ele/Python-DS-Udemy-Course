# Movie Rating Kernel Density Plot

```python
# Imorting the necessary libraries

import pandas as pd 
import matplotlib as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
# Manipulation dataset as previous section

movies = pd.read_csv('Movie-Ratings.csv')
movies.columns = ['Film', 'Genre','CriticRating', 'AudienceRatings'\
              ,'BudgetMillions','Year']
movies.Film = movies.Film.astype('category')
movies.Genre = movies.Genre.astype('category')
movies.Year = movies.Year.astype('category')
movies.head()
```

```python
k1 =  sns.kdeplot(data= movies , x= 'CriticRating', y= 'AudienceRatings'\
                ,shade= True , shade_lowest= False , cmap = "Reds")
#TIPS : To make it looks better

k1b =  sns.kdeplot(data= movies , x= 'CriticRating', y= 'AudienceRatings'\
                , cmap = "Reds")
```
